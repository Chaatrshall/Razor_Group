```python
df = df.replace(np.nan, 'NaN', regex=True)
```

```python
#parsing product counts from sellerproductcount
df['product_count'] = df['sellerproductcount'].str.extract(r'(\d+(?:,\d+)?) results')
```

```python
df.head()
```

```python
#percentage of positive ratings
df['% of ratings'] = df['sellerratings'].str.extract(r'(\d*%) positive')
```

```python
#Number of ratings in last 12 months
df['Ratings-Last 12 months'] = df['sellerratings'].str.extract(r'\((.+)\)')
```

```python
df.head()
```

```python
#parsing phone numbers
df['Contact_Number'] = df['sellerdetails'].str.extract(r"(?:\+|\bTel\.?:|\bFax\.?:|\bTelefon:)\s*(\d+(?:[ ]*[ /-]|[ - /\(0)]*\d+[(0)]*\d+[ /-]*\d+[-]*\d+)*)")
```

```python
#E-mail id of merachants
df['E-mail_ID'] = df['sellerdetails'].str.extract(r"([\w\.-]+@[\w\.-]+)")
```

```python
df.head()
```

```python
df['Location_of_Seller'] = df['businessaddress'].str[-2:]
```

```python
df.head()
```

```python
df.Location_of_Seller.value_counts()
```

```python
df.shape
```

```python
final_df = df[df['Location_of_Seller']!='CN']
```

```python
final_df.shape
```

```python
def f(row):
    
    if row['Hero Product 1 #ratings'] > row['Hero Product 2 #ratings']:
        val = 'Hero Product 1'
    elif row['Hero Product 1 #ratings'] == row['Hero Product 2 #ratings']:
        val = 'Same Ratings'
    else:
        val = 'Hero Product 2'
    return val
```

```python
final_df['Best_Selling_Product'] = final_df.apply(f, axis=1)
```
